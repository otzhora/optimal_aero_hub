# 1 Введение

Узловой аэропорт(хаб) - это аэропорт, использующийся перевозчиками в качестве пересадочной точки для стыковочных рейсов(пассажирских и грузовых), один из современных способов управления транспортными потоками. Пассажиры благодаря "хабу" получают: больше прямых рейсов, больше возможностей вернуться обратным рейсом в тот же день, больший выбор пунктов назначения. Авиакомпаниям "хаб" дает такие преимущества, как: увеличение частоты полетов, расширение сети полетов со сниженной стоимостью. Аэропорты, становясь узловыми, получают возможность: увеличивать пассажиропоток, увеличивать доходы.


# 2 Постановка задачи

Рассмотрим задачу поиска оптимальной промежуточной посадки на маршруте Южная Америка - Северная Америка. При построении модели для исследования будем учитывать такие параметры, как: изменчивость пассажиропотока, дальность самолетов. Также будем использовать такую метрику качества, как: экономия топлива.

При поиске аэропортов, способных выступить в качестве узловых, учтем, что пассажиропотоки со стороны Южной Америки и со стороны Северной Америки для них должны быть примерно одинаковыми.

Для исследования такой метрики качества, как: экономия топлива, составим функцию зависимости стоимости перелета от дальности: $\sum^n_{i=1} p_i * e^{c * d(h, a_i)}$, где $d(h, a_i)$ - расстояние между хабом $h$ и аэропортом $a_i$, $p_i$ - величина пассажиропотока, $c$ - константа, $n$ - количество аэропортов, и минимизируем её по $h$.

Для минимизации функции зависимости стоимости перелета от дальности: $f(x)=\sum^n_{i=1} p_i * e^{c * d(h, a_i)}$ будем применять метод градиентного спуска с критерием остановки $\nabla {f(h_{step-1})} \cdot \nabla {f(h_{step})} < -0.5$. Значение шага выберем равным $50$ м, т.е. на каждом шаге делаем смещение на столько градусов, сколько дают при переводе в десятичные градусы $50$ м. Начальную точку для заполнения массива с координатами "хаба"выбираем случайно. Коэффициент $c$ выбираем равным $1$, $0.5$, $0.25$.

В ходе работы программы при $c = 1$ мы получили точку, находящуюся на территории Никарагуа. От варьирования константы c координаты хаба смещаются, что влияет на выбор ближайшего к ним аэропорта. Ближайший аэропорт, находящийся рядом с полученной точкой при $c = 1$ - Cancún International Airport в Мексике. При $c = 0.5$ точка располагается на территории Кубы и ближайший аэропорт опять - Cancún International Airport в Мексике. При $c = 0.25$ место расположения хаба находится на территории Флориды в США и ближайший аэропорт - Tampa International Airport.
